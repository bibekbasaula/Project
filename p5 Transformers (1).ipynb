Text Classification

```python
from transformers import pipeline
```

```python
pipe = pipeline('text-classification', model = 'tabularisai/multilingual-sentiment-analysis')
```

```python
sentence = '''This captivating mystery, full of twists and turns, kept me on the edge of my seat until the very end'''
```

```python
result = pipe(sentence)
```

```python
result
```

Text Generation

```python
text_generator = pipeline('text-generation', do_sample=True)
```

```python
sentence = 'Why does earth rotate'
```

```python
result = text_generator(sentence)
```
