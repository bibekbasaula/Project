```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Input
```

```python
db = pd.read_csv('/content/drive/MyDrive/broadway py w ai/drug200.csv')
db.head()
```

```python
db.dropna(inplace=True)
```

```python
db.head()
```

```python
db['BP'].unique()
```

```python
db['Cholesterol'].unique()
```

```python
db['Sex'] = db['Sex'].map({'M':0,'F':1})
```

```python
db['BP'] = db['BP'].map({'HIGH':0, 'LOW':1, 'NORMAL':2})
```

```python
db['Cholesterol'] = db['Cholesterol'].map({'HIGH':0, 'NORMAL':1})
```

```python
db
```

```python
db
```

```python
db['Drug'].unique()
```

```python
db['Drug'] = db['Drug'].map({'DrugY':0, 'drugC':1, 'drugX':2, 'drugA':3, 'drugB':4})
```

```python
db
```

```python
x = db[['Age','Sex','BP','Cholesterol','Na_to_K']]
y = db['Drug']
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x , y, test_size = 0.2)
```

```python
model = Sequential([
    Input(shape=(5,)),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(8, activation='relu'),
    Dense(5, activation='softmax')
])
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
fit_model = model.fit(xtrain, ytrain, epochs=75)
```

```python
model.evaluate(xtest, ytest)
```

```python
inp = [[age1,sex1,BP1,cholesterol1,na_to_k1]]
```
