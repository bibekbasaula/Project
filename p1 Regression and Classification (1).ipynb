```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import matplotlib.pyplot as plt
```

```python
db = pd.read_csv('/content/drive/MyDrive/broadway py w ai/Student Performance Multiple Linear Regression.csv')
db.head()
```

```python
db.dropna(inplace=True)
```

```python
x = db[['Hours Studied','Previous Scores','Sleep Hours','Sample Question Papers Practiced']]
y = db['Performance Index']
```

```python
x.head()
```

```python
scaler = StandardScaler()
x=scaler.fit_transform(x)
```

```python
x
```

```python
v = db['Extracurricular Activities']
v.head()
```

```python
for i in range(len(v)):
  if v[i] == 'Yes':
    v[i] = 1
  else:
    v[i] = 0
```

```python
x = pd.DataFrame(x)
```

```python
x[4]=v
```

```python
x.head()
```

```python
y.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Input
```

```python
model = Sequential([
    Input(shape=(5,)),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(8, activation='relu'),
    Dense(4, activation='relu'),
    Dense(1)

])
```

```python
model.compile(optimizer='adam', loss='mae')
```

```python
fmodel = model.fit(xtrain, ytrain, epochs=15)
```

```python
model.evaluate(xtest, ytest)
```
